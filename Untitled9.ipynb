```python
import os

base_dir = os.path.join('rockpaperscissors/rps-cv-images')
print(os.listdir(base_dir))
```

```python
rock_dir = os.path.join(base_dir,'rock')
paper_dir = os.path.join(base_dir, 'paper')
scissors_dir = os.path.join(base_dir,'scissors')

print('rock :', len(os.listdir(rock_dir)))
print('paper: ', len(os.listdir(paper_dir)))
print('scissors:', len(os.listdir(scissors_dir)))
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator (
          rescale = 1.0/255,
          rotation_range = 20,
          horizontal_flip =True,
          shear_range = 0.2,
          zoom_range = 0.2,
          fill_mode = 'nearest',
          validation_split = 0.4)
train_generator = datagen.flow_from_directory(base_dir,
                                       batch_size = 32,
                                       class_mode= 'categorical',
                                       target_size =(150,150),
                                       subset = 'training')
validation_generator= datagen.flow_from_directory(base_dir,
                                            batch_size = 32,
                                            class_mode = 'categorical',
                                            target_size=(150,150),
                                            subset= 'validation')
validationGen = ImageDataGenerator(rescale = 1.0/255,
                                   validation_split = 0.4)
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    
    tf.keras.layers.Dense(3, activation='softmax')])
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer=tf.optimizers.Adam(),
              metrics=['accuracy'])
```

```python
history= model.fit(train_generator,
                   steps_per_epoch = 1314//32,
                   epochs = 10,
                   validation_data=validation_generator,
                   validation_steps= 874//32,
                   shuffle =True,
                   verbose =1)
```

```python
import matplotlib.pyplot as plt


acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training dan Validation accuracy')
plt.legend()
plt.figure()

plt.plot(epochs, loss, 'r', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training dan Validation loss')
plt.legend()

plt.show()
```

```python
import requests
from io import BytesIO
from PIL import Image
import numpy as np

input_size = (150,150) 

channel = (3,0)
input_shape = input_size + channel

labels = ['paper','rock','scissors']
```

```python
def preprocess(img,input_size):
    data_img = img.convert('RGB').resize(input_size, resample= 0)
    img_arr = (np.array(data_img))/255
    return img_arr

def reshape(img1_arr):
    return np.stack(img1_arr, axis=0)
```

```python

img = Image.open('/content/rockpaperscissors/rps-cv-images/paper/0eqArS2GgsBeqgSn.png')
imgplot=plt.imshow(img)
X = preprocess(img, input_size)
X = reshape([X])
y = model.predict(X)
print(labels[np.argmax(y)], np.max(y))
```
